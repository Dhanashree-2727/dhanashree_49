```python
def validate_password(password):
    if len(password) < 8:
        return False

    has_upper = False
    has_lower = False
    has_digit = False

    for ch in password:
        if ch.isupper():
            has_upper = True
        elif ch.islower():
            has_lower = True
        elif ch.isdigit():
            has_digit = True

    return has_upper and has_lower and has_digit


# Test
password = input("Enter password: ")

if validate_password(password):
    print("Valid Password")
else:
    print("Invalid Password")
```
